# Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
# Generate synthetic data
np.random.seed(42)
ages = np.random.rand(100, 1) * 10  # Car ages between 0 and 10 years
values = 50 - 5 * ages + 0.8 * ages**2 + np.random.randn(100, 1) * 2  # Non-linear relationship
```

```python
# Polynomial feature transformation using sklearn
degree = 4  # Higher degree polynomial
poly_reg_model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly_reg_model.fit(ages, values)
```

```python
# Generate predictions
ages_new = np.linspace(0, 10, 100).reshape(-1, 1)
values_predict = poly_reg_model.predict(ages_new)
```

```python
# Plot the data and the polynomial regression curve
plt.figure(figsize=(10, 6))
plt.scatter(ages, values, color="blue", label="Data points")
plt.plot(ages_new, values_predict, color="red", label=f"Polynomial degree {degree}")
plt.xlabel("Age (years)")
plt.ylabel("Value (thousands of dollars)")
plt.title("Polynomial Regression (Degree 4): Predicting Car Value Based on Age")
plt.legend()
plt.grid(True)
plt.show()
```
