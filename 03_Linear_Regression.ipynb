# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate new synthetic data for house sizes (in square feet) and prices (in $1000)
np.random.seed(420) # Change seed for different data generation
sizes = np.random.rand(100, 1) * 3000 + 1000  # Sizes between 1000 and 4000 sqft
prices = 50 + 0.2 * sizes + np.random.rand(100, 1) * 20 # Prices in $1000 with some noise
```

```python
# Fit the linear regression model
model = LinearRegression()
model.fit(sizes, prices)
```

```python
# Generate predictions for the range of sizes
sizes_new = np.linspace(1000, 4000, 100).reshape(-1, 1)
prices_predict = model.predict(sizes_new)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(sizes, prices, color="blue", label="Data points")
plt.plot(sizes_new, prices_predict, color="red", label="Regression line")
plt.xlabel("Size (sqft)")
plt.ylabel("Price (thousands of dollars)")
plt.title("Linear Regression: Predicting House Prices Based on Size")
plt.legend()
plt.grid(True)
plt.show()
```
