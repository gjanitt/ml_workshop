# Principal Component Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)
y = iris.target  # Target (class labels: Setosa, Versicolour, Virginica)
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA to reduce to the top 3 principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Print the explained variance ratio for the top 3 components
print("Explained variance ratio of top 2 principal components:", pca.explained_variance_ratio_)

# Print the top 2 principal components
print("\nTop 2 principal components (PCA components):\n", pca.components_)
```

```python
# Visualizing the data in the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset (Top 2 Principal Components)')
plt.colorbar()
plt.show()
```
